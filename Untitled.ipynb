```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
ex = pd.read_excel("./WSR2.xlsx")
```

```python
currentModel = ex.iloc[0,-1]
modelArray = []
for i, row in ex.iterrows():
    week3 = row['Week 3']
    if (week3 == 'MPP Code:'):
        currentModel = row[0] + ' MPP ' + str(row['Week 5'])
    modelArray.append(currentModel)
```

```python
ex.columns = np.append(['Metric'], ex.columns[1:])
ex['Model'] = modelArray
ex.set_index(ex['Model'], inplace=True)
ex = ex[ex.columns[:-1]]
ex = ex.dropna(subset=ex.columns[-4:])
```

```python
def showModel(modelName):
    df = ex.loc[modelName]
    df = df[np.append('Metric', weeks)]
    df[ df['Metric'].isin(metrics)]
    display(df[ df['Metric'].isin(metrics)])
```

## Seleccionamos las semanas

```python
weeks = ['Week 1', 'Week 2', 'Week 3']
```

## Seleccionamos las m√©tricas

```python
metrics = ['Backorders - Units', 'Total(Inventory) - Units', 'Total(Inventory) - TPC $']
```

## Prueba con pocos elementos

```python
testArray = np.unique(modelArray)[0:9]
```

```python
for modelName in testArray:
    showModel(modelName)
```
